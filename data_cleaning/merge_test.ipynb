```python
import pandas as pd
```

```python
restaurant = pd.read_csv('../clean_data/restaurant.csv')
restaurant = restaurant.rename(columns={'Restaurant Name':'name'})
```

```python
yelp = pd.read_csv('../clean_data/yelp_rating.csv')
```

```python
clean_restaurant= restaurant.drop_duplicates(subset=['name'], keep='first')
clean_restaurant
```

```python
clean_yelp = yelp.drop_duplicates(subset=['name'], keep='first')
clean_yelp
```

```python
combined_df = pd.merge(clean_restaurant, clean_yelp, on='name', how='inner')
combined_df
```
